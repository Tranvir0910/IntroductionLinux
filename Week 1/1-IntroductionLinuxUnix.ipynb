### **Introduction to Linux Unix**

#### **What is an OS**

Đó là phần mềm **Quản lý phần cứng máy tính và tài nguyên Và cho phép tương tác với phần cứng để thực hiện các nhiệm vụ hữu ích**.

#### **What is Unix**

- Unix là một **họ các hệ điều hành.**

- Các hệ điều hành dựa trên Unix phổ biến bao gồm: **Oracle Solaris (và Open Solaris), FreeBSD, HP-UX, IBM AIX và Apple macOS - một trong những hệ điều hành desktop phổ biến nhất ngày nay**.

#### **Unix beginings**

- Unix, ban đầu tạo ra tại AT&T Bell Labs vào những năm 1960, sau đó được viết lại bằng ngôn ngữ lập trình C để trở nên có thể chuyển đổi trên nhiều kiến trúc phần cứng.

- Linux, một họ các hệ điều hành giống Unix, được phát triển như một phiên bản mã nguồn mở và miễn phí của hệ điều hành Unix.

#### **What is Linux**

- Linux là **mã nguồn mở và miễn phí, cho phép bất kỳ ai xem mã nguồn**.

- Hỗ trợ **đa người dùng, chạy đồng thời nhiều người dùng truy cập hệ thống**.

- Hỗ trợ **đa nhiệm, chạy đồng thời nhiều công việc và ứng dụng**.

- Có thể di động, có thể **chạy trên nhiều thiết bị và nền tảng phần cứng khác nhau từ máy tính để bàn đến máy chủ và thiết bị gia đình**.

#### **Linux beginings**

- Trong những năm 1980, dự án GNU được phát triển tại MIT nhằm tạo ra bộ công cụ hệ thống Unix miễn phí và mã nguồn mở.

- Năm 1991, Linus Torvalds phát triển kernel Unix miễn phí và mã nguồn mở được gọi là Linux.

- Trong thập kỷ sau, GNU và kernel Linux được kết hợp để tạo ra các hệ điều hành Linux phổ biến.

- Năm 1996, nhà khoa học máy tính Larry Ewing tạo ra "Tux" con chim cánh cụt, biểu tượng chính thức của Linux.


#### **Linux use cases today**

- Linux được sử dụng rộng rãi trong tỷ lệ tỷ số lượng lớn điện thoại thông minh trên thế giới thông qua hệ điều hành Android, sử dụng kernel dựa trên Linux.

- Linux được sử dụng trong siêu máy tính, nơi hàng triệu máy chủ chạy Linux được kết hợp để tính toán hiệu suất cao.

- Trung tâm dữ liệu doanh nghiệp và đám mây cũng sử dụng Linux trên hàng triệu máy chủ chạy các ứng dụng, máy chủ web, cơ sở dữ liệu, v.v.

- Phiên bản của Linux có thể dễ dàng được sử dụng như hệ điều hành trên máy tính cá nhân.

#### **Today**

- Ngày nay, macOS dựa trên BSD chạy trên hàng triệu thiết bị trên khắp thế giới.

- Tỷ lệ tỷ số lượng lớn các phiên bản Linux chạy trên máy chủ, mang lại trải nghiệm web hiện đại.

- Đặc biệt là trong cộng đồng phát triển, các hệ điều hành Linux hiện đại như Ubuntu đang bắt đầu được ưa chuộng trong không gian máy tính cá nhân.
