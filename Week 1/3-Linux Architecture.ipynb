### **Linux Architecture**

#### **Five layers of Linux**

Hệ thống Linux bao gồm năm tầng khác nhau. 

**UI** : 

- Tầng ngoại cùng của kiến trúc Linux là UI, hoặc giao diện người dùng, cho phép người dùng tương tác với hệ thống bằng cách sử dụng bàn phím hoặc chuột. 

**Application**: 

- Tầng ứng dụng bao gồm các tiến trình hệ thống, shell, ứng dụng người dùng và công cụ được sử dụng để thực hiện các nhiệm vụ trong hệ thống Linux. 

- Các ứng dụng tương tác với hệ điều hành để thực hiện các nhiệm vụ. 

**OPERATING SYSTEM**: 

- Hệ điều hành chịu trách nhiệm cho các công việc quan trọng đối với sự ổn định của hệ thống như lập lịch công việc và theo dõi thời gian. 

- Tất cả các hệ điều hành Linux được xây dựng trên kernel Linux, thực hiện các công việc ở tầng thấp nhất. 

**Kernel**:

- Kernel là thành phần cốt lõi của hệ điều hành và chịu trách nhiệm quản lý bộ nhớ, xử lý và bảo mật. 

**Hardware**

- Kernel tương tác với tầng phần cứng, bao gồm tất cả các thiết bị vật lý hoặc điện tử trong máy tính như bộ xử lý, bộ nhớ, thiết bị nhập và lưu trữ.

**User Interface UI**

**Tầng đầu tiên trong hệ thống Linux là giao diện người dùng**. 

- Giao diện người dùng cho phép người dùng thực hiện các nhiệm vụ bằng cách điều khiển ứng dụng bằng bàn phím. 

- Phiên bản máy tính Linux bao gồm một Giao diện Người Dùng Đồ Họa, hoặc GUI, tương tự như Microsoft Windows.

- Điều này mở rộng chức năng UI đến các thiết bị kiểm soát khác như chuột. 

- Bạn có thể sử dụng máy tính Linux để: 

    - Sử dụng trình duyệt web để gửi email cho bạn bè. 
    
    - Hoặc sử dụng trình phát nhạc để nghe bài hát yêu thích của bạn.

#### **Applications**

- Các ứng dụng là phần mềm cho phép người dùng thực hiện một nhiệm vụ trong hệ thống Linux. 

- Các ứng dụng có thể bao gồm: **Các công cụ hệ thống, như trình biên dịch. Ngôn ngữ lập trình**.

- **Shell**, là các **ứng dụng đặc biệt thường là một phần của hệ điều hành**. 

- Và ứng dụng người dùng, có thể là bất kỳ loại ứng dụng nào, từ trình duyệt, đến trình soạn thảo văn bản, đến trò chơi. 

#### **Operating System**

- **Hệ điều hành kiểm soát các công việc và chương trình quan trọng đối với sức khỏe và ổn định của hệ thống**. 

- Các chức năng của nó cũng bao gồm: **Gán phần mềm cho người dùng**, **Phát hiện lỗi** và **triển khai biện pháp để ngăn chặn sự cố hệ thống hoàn toàn, Và thực hiện quản lý tệp tin**. 

- Trong hệ thống Linux, hệ điều hành được **xây dựng trên kernel Linux**, **thực hiện các thao tác quan trọng nhất**.

#### **Kernel**

- Kernel là **phần mềm ở tầng thấp nhất trong hệ thống Linux và có quyền kiểm soát hoàn toàn hệ thống**. 

- Nó **khởi động ngay khi máy tính của bạn khởi động** và **vẫn ở trong bộ nhớ của máy tính của bạn trong khi hệ thống đang chạy**. 

- Nó cũng **hoạt động như một cầu nối giữa ứng dụng ( Application ) của bạn và phần cứng máy tính ( hardware ) của bạn**, cho phép **hai bên tương tác hiệu quả sử dụng "system calls"**. 

- Kernel có bốn công việc chính: 

    - Quản lý bộ nhớ, 

    - Quản lý quy trình, 

    - Quản lý trình điều khiển thiết bị để hỗ trợ phần cứng đúng cách

    - Đảm bảo hệ thống duy trì tính bảo mật.

#### **Hardware**

- **Tầng cuối cùng trong hệ thống Linux là phần cứng**, bao gồm **tất cả các thành phần vật lý hoặc điện tử của máy tính**. 

- Tầng này có thể bao gồm các thành phần sau: 

    - **Bộ xử lý trung tâm, hoặc CPU**, chịu trách nhiệm **thực hiện hầu hết các phép tính**. 

    - **Bộ nhớ truy cập ngẫu nhiên, hoặc RAM**, là **một đơn vị lưu trữ nhanh được sử dụng để giữ thông tin tạm thời mà ứng dụng của bạn cần chạy**. 

    - **Bộ nhớ lưu trữ cho dữ liệu** cần tồn tại khi máy tính của bạn tắt nguồn. Màn hình máy tính của bạn. Và bất kỳ thiết bị USB nào như bàn phím, chuột hoặc ổ USB.

#### **Linux Filesystem**

- Hệ thống tệp Linux là **bộ sưu tập các tệp tin trên máy tính của bạn**.

- Nó bao gồm **các tệp tin cần thiết để chạy máy** và **các ứng dụng cũng như các tệp cá nhân của bạn chứa công việc của bạn**.

- **Tầng cao nhất của hệ thống tệp là thư mục gốc**, được **biểu tượng bằng dấu gạch chéo chéo (/)**.

- Dưới đây là một cấu trúc giống cây của các thư mục và tệp tin trong hệ thống. Và hệ thống tệp gán quyền truy cập phù hợp cho các thư mục và tệp tin:

    - Phần trên cùng của hệ thống tệp Linux là thư mục gốc, chứa nhiều thư mục và tệp tin khác nhau.

    - Một trong những **thư mục chính là /bin**, **chứa các tệp nhị phân người dùng**.

        - Tệp nhị phân **chứa mã mà máy tính của bạn đọc để chạy chương trình và thực hiện các lệnh**.

        - Nó được gọi là **"slash bin"** để **chỉ ra rằng nó tồn tại trực tiếp dưới thư mục gốc**.

    - Các **thư mục chính khác** bao gồm: 
    
        - **/usr**, chứa **các chương trình người dùng**.

        - **/home** là **thư mục làm việc cá nhân** của bạn nơi bạn **nên lưu trữ tất cả các tệp cá nhân của mình**.

        - **/boot** chứa **các tệp khởi động hệ thống** của bạn, **hướng dẫn quan trọng cho khởi động hệ thống**.

        - **/media** chứa **các tệp liên quan đến phương tiện** tạm thời như CD hoặc ổ đĩa USB kết nối với hệ thống.

        - Còn nhiều thư mục khác trong thư mục gốc, nhưng bạn sẽ không cần truy cập chúng trong khoá học này.

- Tất cả các tệp và thư mục trong hệ thống Linux được tổ chức vào một trong những thư mục được chỉ định này, tùy thuộc vào mục đích của tệp hoặc thư mục.

#### **Recap**

1. **Hệ thống Linux bao gồm năm tầng chính:**
   - Giao diện người dùng
   - Ứng dụng
   - Hệ điều hành
   - Kernel
   - Phần cứng

2. **Giao diện người dùng là tầng cho phép người dùng tương tác với ứng dụng sử dụng thiết bị kiểm soát:**
   - Giao diện này cung cấp phương tiện cho người dùng để tương tác với hệ thống bằng cách sử dụng bàn phím, chuột hoặc các thiết bị kiểm soát khác.

3. **Các ứng dụng cho phép người dùng thực hiện một số nhiệm vụ cụ thể trong hệ thống:**
   - Ứng dụng bao gồm các công cụ hệ thống, ngôn ngữ lập trình, các shell, và ứng dụng người dùng như trình duyệt web, trình phát nhạc, hoặc trình soạn thảo văn bản.

4. **Hệ điều hành chạy trên kernel Linux và quan trọng cho sức khỏe và ổn định của hệ thống:**
   - Hệ điều hành thực hiện các công việc quan trọng như lập lịch công việc và theo dõi thời gian, đồng thời quản lý các tác vụ quan trọng khác.

5. **Kernel là phần mềm ở tầng thấp nhất và cho phép các ứng dụng tương tác với phần cứng của bạn:**
   - Kernel quản lý bộ nhớ, quản lý quy trình, quản lý các trình điều khiển thiết bị cho hỗ trợ phần cứng, và đảm bảo an ninh hệ thống.

6. **Phần cứng bao gồm tất cả các thành phần vật lý hoặc điện tử của máy tính của bạn:**
   - Bao gồm CPU (trung tâm xử lý), RAM (bộ nhớ truy cập ngẫu nhiên), lưu trữ dữ liệu, màn hình máy tính, và các thiết bị ngoại vi như bàn phím, chuột, USB, và nhiều hơn nữa.

7. **Và hệ thống tệp Linux là một cấu trúc giống cây bao gồm tất cả các thư mục và tệp trên hệ thống:**
   - Hệ thống tệp bắt đầu từ thư mục gốc (/) và có cấu trúc dạng cây, với nhiều thư mục và tệp tin phân bố theo mục đích cụ thể, như /bin cho các tệp nhị phân người dùng, /usr cho các chương trình người dùng, /home cho thư mục cá nhân của người dùng, và nhiều thư mục khác.
