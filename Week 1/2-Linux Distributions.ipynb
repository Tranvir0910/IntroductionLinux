### **Linux Distributions**

#### **What is a Linux distribution**

- Mt Linux Distribution l **mt bin th c th ca h iu hnh Linux. N cng c th c gi l "distro"**.

#### **Linux distro differences**

- Mi Linux Distribution bao gm **mt b tin ch mc nh duy nht l cc lnh v ng dng c ng gi trc vi phn phi **.

- Mi distro **c giao din ngi dng  ha (GUI) ring bit  tng tc vi h iu hnh**.

- Mi distro **h tr mt b lnh c th m bn c th s dng trong shell (mt ca s  nhp v nhn u ra t cc lnh)**.

- Mi distro **cung cp cc mc  h tr khc nhau, c th c pht trin v qun l nh mt d n ca cng ng hoc c duy tr bi doanh nghip thng mi**. 

- N c th l phin bn **h tr di hn (LTS) hoc mt pht hnh lin tc (rolling release) vi cp nht n nh theo lch trnh lin tc**.

#### **Linux distros: Debian**

**Debian: Mt trong nhng distro c ngun gc lu di, n nh, ng tin cy v hon ton m ngun m. Ph bin trong khng gian my ch.**

#### **Ubuntu**

- Ubuntu: Da trn Debian, c pht trin bi Canonical. 

- C ba phin bn chnh: **Desktop (cho my tnh c nhn), Server (dnh cho my ch n gin hoc m my nhiu nt) v Core (cho Internet of Things)**.

#### **Red Hat Linux**

- Red Hat Linux (RHEL): **Mt distro "core" khng phi l pht sinh t distro Linux khc, c bit n vi tnh n nh, ng tin cy v hon ton m ngun m.**

- c qun l bi Red Hat, mt cng ty con ca IBM, v **tp trung hon ton vo khch hng doanh nghip**.

#### **Fedora**

- Fedora: **Stabil, h tr nhiu kin trc, an ton v ng tin cy, cung cp tnh nng bo mt tng la v duy tr mt cng ng ln ang pht trin**. c ti tr bi Red Hat.

#### **SUSE Enterpise**

- SUSE Linux Enterprise (SLE): **C hai phin bn: Server (SLES) v Desktop (SLED)**. 

- H tr nhiu kin trc, bao gm **ARM cho Raspberry Pi. c duy tr bi cng ty phn mm m ngun m nh c, SUSE**.

#### **Arch Linux**

- Arch Linux: **Tip cn "lm nn bn t" cho php ngi dng ty chnh mi phn ca h thng.** 

- Rt linh hot, nhng **i hi s hiu bit mnh m v Linux v cc cng c h thng  s dng Arch hiu qu.** 

- Cung cp **quyn t do ln nhng cng c ngha l bn cn hiu bit su rng v Linux  s dng Arch mt cch hiu qu**.

#### **Recap**

- Thit k ca mt Linux Distribution c ti u ha cho mt i tng ngi dng c th. 

- Debian c coi l n nh v ng tin cy cho khng gian my ch, trong khi RHEL tp trung hon ton vo doanh nghip. 

- SUSE Linux Enterprise h tr nhiu kin trc v Arch Linux mang li s ty chnh cao nhng i hi s hiu bit su rng.
