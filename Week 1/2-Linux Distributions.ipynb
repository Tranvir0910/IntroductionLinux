### **Linux Distributions**

#### **What is a Linux distribution**

- Một Linux Distribution là **một biến thể cụ thể của hệ điều hành Linux. Nó cũng có thể được gọi là "distro"**.

#### **Linux distro differences**

- Mỗi Linux Distribution bao gồm **một bộ tiện ích mặc định duy nhất là các lệnh và ứng dụng được đóng gói trước với phân phối đó**.

- Mỗi distro **có giao diện người dùng đồ họa (GUI) riêng biệt để tương tác với hệ điều hành**.

- Mỗi distro **hỗ trợ một bộ lệnh cụ thể mà bạn có thể sử dụng trong shell (một cửa sổ để nhập và nhận đầu ra từ các lệnh)**.

- Mỗi distro **cung cấp các mức độ hỗ trợ khác nhau, có thể được phát triển và quản lý như một dự án của cộng đồng hoặc được duy trì bởi doanh nghiệp thương mại**. 

- Nó có thể là phiên bản **hỗ trợ dài hạn (LTS) hoặc một phát hành liên tục (rolling release) với cập nhật ổn định theo lịch trình liên tục**.

#### **Linux distros: Debian**

**Debian: Một trong những distro có nguồn gốc lâu dài, ổn định, đáng tin cậy và hoàn toàn mã nguồn mở. Phổ biến trong không gian máy chủ.**

#### **Ubuntu**

- Ubuntu: Dựa trên Debian, được phát triển bởi Canonical. 

- Có ba phiên bản chính: **Desktop (cho máy tính cá nhân), Server (dành cho máy chủ đơn giản hoặc đám mây nhiều nút) và Core (cho Internet of Things)**.

#### **Red Hat Linux**

- Red Hat Linux (RHEL): **Một distro "core" không phải là phát sinh từ distro Linux khác, được biết đến với tính ổn định, đáng tin cậy và hoàn toàn mã nguồn mở.**

- Được quản lý bởi Red Hat, một công ty con của IBM, và **tập trung hoàn toàn vào khách hàng doanh nghiệp**.

#### **Fedora**

- Fedora: **Stabil, hỗ trợ nhiều kiến trúc, an toàn và đáng tin cậy, cung cấp tính năng bảo mật tường lửa và duy trì một cộng đồng lớn đang phát triển**. Được tài trợ bởi Red Hat.

#### **SUSE Enterpise**

- SUSE Linux Enterprise (SLE): **Có hai phiên bản: Server (SLES) và Desktop (SLED)**. 

- Hỗ trợ nhiều kiến trúc, bao gồm **ARM cho Raspberry Pi. Được duy trì bởi công ty phần mềm mã nguồn mở nhỏ Đức, SUSE**.

#### **Arch Linux**

- Arch Linux: **Tiếp cận "làm nên bạn tự" cho phép người dùng tùy chỉnh mọi phần của hệ thống.** 

- Rất linh hoạt, nhưng **đòi hỏi sự hiểu biết mạnh mẽ về Linux và các công cụ hệ thống để sử dụng Arch hiệu quả.** 

- Cung cấp **quyền tự do lớn nhưng cũng có nghĩa là bạn cần hiểu biết sâu rộng về Linux để sử dụng Arch một cách hiệu quả**.

#### **Recap**

- Thiết kế của một Linux Distribution được tối ưu hóa cho một đối tượng người dùng cụ thể. 

- Debian được coi là ổn định và đáng tin cậy cho không gian máy chủ, trong khi RHEL tập trung hoàn toàn vào doanh nghiệp. 

- SUSE Linux Enterprise hỗ trợ nhiều kiến trúc và Arch Linux mang lại sự tùy chỉnh cao nhưng đòi hỏi sự hiểu biết sâu rộng.
