### **Linux Terminal Overview**

#### **Overview of the Linux shell**

- **Shell Linux** là **một ứng dụng cấp hệ điều hành giải thích các lệnh**. 

- Trong những phiên bản sớm của Unix và Linux, **shell là cách duy nhất để tương tác với hệ điều hành**. 

- Ngày nay, bạn cũng **có thể sử dụng giao diện đồ họa**, nhưng **shell vẫn là sự lựa chọn phổ biến** và **linh hoạt cũng như một cách dễ dàng để chạy các tệp script**. 

- Bạn có thể **sử dụng lệnh shell** để:

    - Di chuyển và sao chép tệp. 

    - Viết vào và đọc từ tệp.

    - Trích xuất và lọc dữ liệu.

    - Tìm kiếm dữ liệu. 

- Có **nhiều phiên bản shell khác nhau**, nhưng **chức năng cơ bản của hầu hết đều giống nhau**. Một số ví dụ phổ biến bao gồm: **Bash và Zsh**.

#### **Overview of a Linux Terminal**

- Bạn **tương tác với shell Linux thông qua terminal Linux**. 

- Một terminal là **một ứng dụng hoặc giao diện người dùng**, nơi **bạn nhập các lệnh bạn muốn chạy** và **nhận bất kỳ đầu ra nào từ những lệnh đó**. 

- Ví dụ, để khởi **chạy ứng dụng Python** và chạy một **chương trình có tên là "myprogram.py"**, nhập ***python space myprogram.py***. 

- Khi bạn nhấn Enter, shell chạy lệnh. Chương trình này in ra các từ ***"Hello, World!" trên terminal***.

#### **Communicating with Linux system**

- Họ **nhập lệnh vào một terminal, sau đó được chuyển đến shell**. 

- Các **thành phần cốt lõi của hệ điều hành** và **kernel dịch lệnh để phần cứng thực hiện**. 

- Khi **phần cứng hoàn tất lệnh**, **kernel đọc bất kỳ thay đổi hoặc kết quả nào** và **gửi chúng trở lại qua shell** **đến terminal** để **thông báo cho người dùng**. 

- **Terminal là một cách mạnh mẽ để chạy ứng dụng và tương tác với máy của bạn**.

#### **The Linux Terminal**

- Hầu hết **các terminal có một giao diện người dùng tương tự để bạn nhập lệnh**. **Khu vực nơi bạn nhập lệnh được gọi là dòng lệnh**.  

- **Và dòng thẳng | , hoặc con trỏ, là dấu nhắc lệnh**. Điều này **chỉ ra nơi văn bản mà bạn nhập sẽ được hiển thị**.

- Trong ví dụ này, thư mục làm việc hiện tại là thư mục Documents, nằm trong thư mục me, nằm trong thư mục home. 

- Thư mục làm việc hiện tại là nơi mà shell sẽ tìm kiếm bất kỳ lệnh nào bạn chỉ định để chạy, ví dụ như chương trình Python trong ví dụ trước. Không phải tất cả các terminal hiển thị đầy đủ vị trí, hoặc đường dẫn, của thư mục hiện tại của bạn, vì vậy một số chỉ hiển thị Documents ở đây.

#### **Paths in the Linux filesystem**

- **Không phải tất cả các terminal hiển thị đầy đủ vị trí, hoặc đường dẫn, của thư mục hiện tại của bạn**, vì vậy **một số chỉ hiển thị Documents ở đây**. 

- **Đường dẫn là vị trí có thể đọc được của một thư mục hoặc tệp trong hệ thống tệp Linux**. 

- Cấu trúc **"a/b"** cho biết rằng **tệp hoặc thư mục có tên "b" nằm bên trong thư mục có tên "a"**. 

- **Cũng có các đường dẫn đặc biệt:** 

    - Một **ký hiệu tilde ( ~ )** duy nhất **đề cập đến thư mục home của người dùng**. 
    
    - Một **dấu gạch chéo ( / )** duy nhất **ở đầu đường dẫn đề cập đến thư mục root**. 
    
    - **Hai dấu chấm ( .. ) đề cập đến thư mục cha của thư mục hiện tại**. 
    
    - Và **một dấu chấm đơn ( . ) đề cập đến thư mục hiện tại**. 
    
Vậy nên, hãy xem cách bạn có thể sử dụng terminal để thay đổi thư mục làm việc hiện tại.

#### **Changing the current working directory**

- Bạn sử dụng lệnh **cd có nghĩa là thay đổi thư mục**. 

- Nhập ***cd /*** để **đi đến thư mục root**. 

- Và nhập ***cd bin*** để **chuyển vào thư mục bin**. 

- Thư mục **bin**, nằm **trong thư mục root**, **chứa các chương trình cần thiết cho hệ thống**. 

- Một trong **những chương trình hoặc tệp thực thi này được gọi là "ls"**.

- Bạn có thể **chạy chương trình ls trong thư mục làm việc hiện tại bằng cách nhập ./ls**. 

- Sử dụng nó **để hiển thị trong cửa sổ terminal tên của tất cả các tệp** và **thư mục trong thư mục hiện tại**. 

- **Nhiều lệnh nằm trong thư mục bin cũng được tích hợp vào shell**, vì vậy **bạn có thể chạy chúng từ các vị trí khác nhau**. 

- Di chuyển **đến thư mục home của bạn bằng cách nhập cd ~**. Và mặc dù thư mục làm việc hiện tại, **/home/me**, **không chứa chương trình ls, bạn vẫn có thể chạy lệnh thành công**. 

- Làm thế nào để thay đổi từ **/home directory.** Nhập **"cd .."** để chuyển **đến thư mục cha của thư mục làm việc hiện tại**. 

- Vì vậy, trong ví dụ này, thư mục **cha của /home là /**, hoặc thư mục **root**. 

- Sau đó, để **di chuyển đến một ổ đĩa USB có tên là my-usb-drive trong thư mục media**, nhập ***cd /media/my-usb-drive***.

- Bạn cũng có thể di chuyển lên và xuống cây trong một lệnh.  

- Để **di chuyển lên thư mục media** sau đó **lên thư mục root**, nhập ***cd ../..*** và sau đó để **di chuyển xuống thư mục Documents** trong thư mục **me** trong thư mục **home**, nhập  /**home/me/Documents**. 

- Nhấn Enter để gửi lệnh và **di chuyển đến thư mục Documents**. 

- Hãy **di chuyển lên đến thư mục /home/me**, **khởi chạy ứng dụng Python** và **chạy một chương trình trong thư mục /home/me** có tên là **myprogram.py**.

- Chương trình này trả về một thông báo trên cửa sổ terminal.


#### **RECAP**

- Shell Linux là một ứng dụng cấp hệ điều hành mà bạn có thể sử dụng để nhập lệnh và xem đầu ra của những lệnh đó. 

- Bạn sử dụng terminal để gửi lệnh đến shell. Và bạn có thể sử dụng lệnh cd để di chuyển xung quanh hệ thống tệp Linux của bạn.
