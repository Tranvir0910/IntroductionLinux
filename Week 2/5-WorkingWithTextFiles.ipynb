### **Working with Text Files**

#### **Viewing file contents**

```python
$ cat entrypoint.sh

$ more entrypoint.sh

$ less entrypoint.sh
```


- `cat (catenate)` - In ra toàn bộ tệp tin.

- `more` - In tệp tin theo từng trang.

- `less` - In tệp tin ra và có thể scroll bằng chuột.


```python
$ head usdoi.txt

$ head -3 usdoi.txt

$ tail usdoi.txt

$ tail -2 usdoi.txt
```

- `head` - In ra 10 dòng đầu tiên của tệp.

- `head -N` - In ra N line đầu tiên.

- `tail` - In ra 10 dòng cuối cùng của tệp.

- `tail -N` - In ra N line cuối cùng.

#### **Getting basic text file stats**

- **Count lines, words, or characters from a text file**

```python
$ wc usdoi.txt

$ wc -l usdoi.txt

$ wc -w usdoi.txt

$ wc -c usdoi.txt
```

`wc (word count)` - Đếm số char, word, line.

`wc -l` - Đếm số dòng.

`wc -w` - Đếm số từ.

`wc -c` - Đếm số kí tự.

- **Sort and display lines of file alphanumerically**

```python
$ sort usdoi.txt

$ sort -r usdoi.txt
```

`sort` - sắp xếp các dòng trong file theo thứ tự bảng chử cái.

`sort -r` - sắp xếp theo cách ngược lại. (Reverse)

- **Drop consecutive duplicated lines and display result**

```python
$ uniq zoo.txt
```

`uniq (unique)` - lọc ra các dòng lặp đi lặp lại.

- **Extract lines matching a specified criterion**

```python
$ grep people usdoi.txt
```

`grep ( global regular expression print )` - trích xuất các dòng phù hợp với tiêu chí đã chỉ định

- **Extract fields from lines of text**

```python
$ cut -c -2 zoo.txt

$ cut -c 2- zoo.txt

$ cut -d "," -f2 names_and_numbers.csv
```

`cut` - bằng tùy chọn `-c -2` để chỉ xem hai ký tự đầu tiên của mỗi dòng

Hoặc để xem từng dòng bắt đầu từ ký tự thứ hai `( -c 2- )`

```python
$ cut -d "," -f2 names_and_numbers.csv

          |   |
          |   |
         (1) (2)

(1) - ',' loại bỏ dấu ,
(2) - '-f2' loại bỏ từ đầu trước ',' ( first word )
```

- **Merge text files line-by-line, aligned as columns**

```python
$ paste zoo.txt zoo_ages.txt

$ paste -d "," zoo.txt zoo_ages.txt
```

`paste` - Kết hợp các line từ các tệp khác nhau.

Mặc định khi paste sẽ là dấu space nhưng có thể tùy chọn như sau:

`paste -d "," zoo.txt zoo_ages.txt`
