### **File and folder archiving and compression**

#### **Create and manage file archives**

`tar` - cho phép đóng gói nhiều tệp và thư mục vào một tệp lưu trữ duy nhất.

- Toàn bộ thư mục /bin và ghi kho lưu trữ vào một tệp duy nhất có tên bin.tar.

```python
$ tar -cvf bin.tar /bin
```

- Để xem danh sách các file trong kho lưu trữ, hãy sử dụng tùy chọn `-t`.

- Để gỡ bỏ kho lưu trữ hoặc giải nén các tệp từ kho lưu trữ, hãy sử dụng tùy chọn `-x`

- Sử dụng `ls -l` để biết `dir` đã được giải nén.

#### **Package and compress archive files**

`zip` - cho phép nén tệp tin.

- Lệnh sau tạo một tệp zip có tên `config.zip` bao gồm tất cả các tệp `có phần mở rộng .conf` trong `thư mục /etc`.

```python
$ zip config.zip /etc/*.conf
```

- Tùy chọn `-r` có thể được sử dụng để nén toàn bộ thư mục.

- Lệnh sau tạo một kho lưu trữ của thư mục /bin.

```python
$ zip -r bin.zip /bin
```

#### **Extract, list, or test compressed files in a ZIP archive**

`unzip` - cho phép giải nén tệp tin.

- Để liệt kê các tệp của kho lưu trữ config.zip, hãy nhập như sau:

- Sử dụng `-l`

```python
$ unzip -l config.zip
```

- Trích xuất tất cả các tệp trong kho lưu trữ bin.zip.

```python
$ unzip -o bin.zip
```

- Thêm tùy chọn `-o` để buộc `ghi đè` trong trường hợp `chạy lệnh nhiều lần`.
