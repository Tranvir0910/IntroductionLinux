### **File and Directory Management Commands**

#### **Creating directories**

- Giả sử bạn đang ở trong thư mục tài liệu của mình, hiện đang trống rỗng. 

- Để tạo một thư mục có tên là `"test"`, bạn chỉ cần nhập `"mkdir test"` Bây giờ thư mục làm việc của bạn có một thư mục con có tên là `"test"`. 

#### **Removing files and directories**

- Lệnh `rm` cho phép bạn **xóa một tệp hoặc thư mục**. Giả sử bạn có một cấu trúc tệp như hình bên trái. Và, giả sử bạn đang ở trong thư mục `Documents` và bạn muốn `xóa "file1"`. 

- Bạn có thể làm điều này bằng cách đơn giản nhập `"rm file1"`. Bây giờ bạn chỉ thấy còn `"folder1"`. 

- Bạn **không thể đơn giản xóa "folder1" vì nó có thể chứa các tệp khác**. 

- Tuy nhiên, bạn có thể dễ dàng làm điều này bằng cách nhập `"rm"` với tùy chọn `-r`. Tùy chọn `-r` **có nghĩa là bạn muốn xóa thư mục cùng với tất cả các tệp con của nó**. 

- Thư mục `Documents` của bạn bây giờ trống rỗng. Do đó, bạn nên **luôn cẩn thận khi sử dụng rm với tùy chọn -r**. Rất dễ xóa các thư mục chứa dữ liệu quan trọng một cách vô tình. 

- Bây giờ, giả sử bạn **tạo một thư mục trống bằng lệnh mkdir** và sau đó quyết định xóa nó. 

- Lệnh **rm option rf không được khuyến nghị**. Thay vào đó, bạn nên `sử dụng lệnh rmdir`, được sử dụng duy nhất để xóa các thư mục trống. 

- Điều này đảm bảo rằng bạn sẽ không bao giờ xóa tình cờ bất kỳ tệp hoặc thư mục quan trọng nào. Nhập "ls" sẽ cho thấy rằng thư mục hiện tại thực sự là trống rỗng.

#### **Creating files**

- Lệnh `touch` có thể **được sử dụng để tạo các tệp rỗng**. Giả sử bạn đang ở trong thư mục tài liệu của mình, nơi trống rỗng, và bạn muốn tạo một số tệp văn bản rỗng. 

- Bạn có thể làm điều này bằng cách nhập `"touch"` cùng với một số tên tệp: `"a", "b", "c" và "d" với hậu tố ".txt"`. Bây giờ, bạn có thể thấy rằng thư mục tài liệu của bạn chứa bốn tệp bạn vừa tạo. 

- Bạn có thể đang tự hỏi rằng lệnh touch làm gì với một tệp đã tồn tại. Giả sử thư mục hiện tại của bạn chứa một tệp có tên là "notes.txt". 

- Bạn có thể xem lần cuối cùng nó được sửa đổi bằng `'date -r notes.txt’`. 

- Nếu bạn `sử dụng lệnh touch trên notes.txt`, bạn có thể thấy **ngày sửa đổi lần cuối cùng của nó đã được cập nhật đến thời gian hiện tại**. 

#### **Copying files and directories**

-  Lệnh `cp` **cho phép bạn sao chép một tệp hoặc thư mục**. 

- Để sao chép tệp, bạn có thể hoặc: sao chép một tệp từ một thư mục nguồn và chỉ định tên tệp cho nó trong thư mục đích, hoặc đơn giản là bỏ qua tên tệp đích và mặc định giữ nguyên tên tệp. 

- Để **sao chép toàn bộ thư mục**, bạn cần thêm `tùy chọn -r cho cp` để nó biết sao chép đệ quy tất cả các thư mục và tệp con. 

- Hãy xem một số ví dụ: Giả sử bạn có một tệp có tên là "notes.txt" và một thư mục có tên là "Documents" trong thư mục làm việc của bạn. 

- Bạn có thể **sao chép notes.txt vào thư mục "Documents"** bằng `"cp notes.txt Documents"`. 

- Bây giờ bạn có thể thấy rằng thư mục Documents của bạn chứa một bản sao của notes.txt. 

- Chú ý rằng bạn không cần phải chỉ định một thư mục nguồn, vì cp mặc định là thư mục hiện tại của bạn. 

- Tiếp theo, bạn có thể tạo một bản sao của thư mục Documents của mình, gọi là "Docs_copy", bằng cách sử dụng cú pháp từ bên trái. 

- Và như mong đợi, bây giờ bạn có một thư mục "Docs_copy" chứa cùng nội dung như thư mục Documents gốc.

#### **Moving files and directories**

- Lệnh `mv` **cho phép bạn di chuyển một tệp hoặc thư mục**. Để di chuyển tệp, bạn có thể nhập `"mv"` tiếp theo là **đường dẫn của các tệp bạn muốn di chuyển, tiếp theo là thư mục bạn muốn di chuyển chúng đến**. 

- Tương tự, để di chuyển các thư mục, bạn nhập "mv" với đường dẫn của thư mục cần di chuyển, tiếp theo là đường dẫn và thư mục bạn muốn di chuyển chúng đến. 

- Hãy xem một ví dụ: Giả sử bạn có một tệp có tên là `"my_script.sh"` và ba thư mục, gọi là "Scripts", "Notes" và "Documents". 

- Bạn có thể `di chuyển "my_script.sh" vào thư mục "Scripts"` bằng cú pháp từ bên trái. Tương ứng, nhập `"ls my_script.sh" sẽ không trả về gì cả`, nhưng `nhập "ls Scripts" sẽ cho thấy bạn đã di chuyển my_script.sh thành công vào thư mục Scripts của bạn`. 

- Tiếp theo, bạn có thể di chuyển các thư mục Notes và Scripts vào thư mục Documents của bạn bằng cú pháp ở bên trái. 

- Bạn có thể thấy rằng thư mục của bạn bây giờ chỉ chứa thư mục Documents và thư mục Documents của bạn chứa các thư mục Scripts và Notes bạn vừa di chuyển.

- Sử dụng mv để đổi tên user.txt thành user-info.txt bằng cách nhập lệnh sau:

`mv users.txt user-info.txt`

#### **Managin file permissions**

- `chmod` là viết tắt của `"change mode"` và được **sử dụng để thay đổi quyền đọc, ghi và thực thi trên các tệp**. 

- Giả sử bạn có một tệp lệnh shell trong thư mục hiện tại của bạn có tên là `"my_script.sh"` và in ra `"Learning Linux is fun!"`. 

- Nhập `ls -l my_script.sh` sẽ **cho thấy rằng tệp lệnh của bạn có quyền đọc và ghi, như được chỉ ra bằng các ký tự r và w**. 

- Nhưng nếu bạn cố gắng thực thi tệp, bạn sẽ nhận được lỗi từ chối quyền. 

- Để làm cho tệp lệnh của bạn có thể thực thi, bạn gọi `chmod trên my_script.sh với tùy chọn +x`. 

- Nhập `ls -l my_script.sh` bây giờ sẽ **cho thấy rằng my_script.sh có quyền thực thi, như được chỉ ra bởi ký tự x**.

#### **Recap**

- Sử dụng lệnh touch để tạo một tệp mới hoặc cập nhật ngày cuối cùng mà một tệp đã tồn tại. 

- Tạo thư mục với lệnh mkdir và xóa một thư mục trống với lệnh rmdir. 

- Sử dụng các lệnh cp và mv để sao chép, di chuyển và đổi tên tệp và thư mục.

- Sử dụng chmod để thay đổi quyền đọc, ghi và thực thi trên các tệp.
