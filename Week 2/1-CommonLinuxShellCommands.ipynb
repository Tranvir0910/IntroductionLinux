### **Overview of Common Linux Shell Commands**

#### **What is a shell**

- Shell là **một giao diện người dùng mạnh mẽ cho các hệ điều hành giống Unix**. 

- Nó có thể hiểu lệnh và chạy các chương trình khác. 

- Một shell, **cho phép truy cập vào các tệp tin, tiện ích và ứng dụng, cũng là một ngôn ngữ tương tác**. 

- Shell cũng là **một ngôn ngữ kịch bản** và **có thể được sử dụng để tự động hóa các nhiệm vụ**.

#### **A sea of shells**

- Shell mặc định trên hệ thống Linux thường là Bash. 

- Các shell khác bao gồm `'Bourne shell, SH', 'Korn shell, KSH', 'T C shell', 'Z shell' và 'fish'`. 

- Trong khóa học này, chúng tôi chỉ sử dụng shell Bash, viết tắt của 'bourne again shell'. 

- Để tìm hiểu shell mặc định là gì, nhập `'printenv SHELL'` vào dòng lệnh. Điều này trả về đường dẫn đến chương trình shell mặc định, trong trường hợp này là bash. 

- Nếu shell mặc định của bạn không phải là Bash, bạn có thể chuyển đổi sang nó bằng cách nhập `'bash'` vào dòng lệnh. 

- Chúng tôi sẽ sử dụng dấu `$` để đại diện cho dấu nhắc lệnh trong toàn bộ khóa học này. 

- Bên ngoài khóa học này, bạn có thể gặp biểu tượng `'greater than <'` được sử dụng cho mục đích tương tự.

#### **Shell Command Applications**

- Các ứng dụng của các lệnh shell bao gồm: 

    - Lấy thông tin. 

    - Duyệt và làm việc với tệp tin và thư mục. 

    - In nội dung của tệp tin và chuỗi - Nén và lưu trữ tệp tin. 

    - Thực hiện các thao tác mạng. 

    - Giám sát hiệu suất và trạng thái của hệ thống, các thành phần và ứng dụng của nó.

    - Chạy công việc hàng loạt, như các thao tác ETL.

#### **Getting information**

- Một số lệnh shell phổ biến để lấy thông tin bao gồm: 

    - `whoami` - trả về tên người dùng. 

    - `id` - trả về người dùng hiện tại và ID nhóm. 

    - `uname` - trả về tên hệ điều hành, ps - hiển thị quá trình đang chạy và ID của chúng.

    - `top` - hiển thị quá trình đang chạy và sử dụng tài nguyên bao gồm bộ nhớ, CPU và IO.

    - `df` - hiển thị thông tin về các hệ thống tệp đã được mount.

    - `man` - truy xuất sách hướng dẫn tham chiếu cho bất kỳ lệnh shell nào.

    - `date` - in ra ngày hôm nay.

#### **Working with files**

- Một số lệnh shell phổ biến để làm việc với tệp tin bao gồm: 

    - `cp` - sao chép tệp tin.

    - `mv` - đổi tên tệp tin hoặc đường dẫn.

    - `rm` - xóa tệp tin.

    - `touch` - tạo tệp tin trống, cập nhật dấu thời gian của tệp tin. 

    - `changemod` - thay đổi hoặc sửa đổi quyền truy cập của tệp tin.

    - `wc` - lấy số lượng dòng, từ và ký tự trong tệp tin,

    - `grep` - trả về các dòng trong tệp tin khớp mẫu. 

#### **Navigating and working with directories**

- Các lệnh shell rất phổ biến để duyệt và làm việc với thư mục bao gồm: 

    - `ls` - liệt kê các tệp tin và thư mục trong thư mục hiện tại.
    
    - `find` - được sử dụng để tìm kiếm các tệp tin khớp mẫu trong cây thư mục hiện tại.
    
    - `pwd` - in ra thư mục hiện tại, hoặc `'present working'`.
    
    - `mkdir` - tạo một thư mục mới.
    
    - `cd` - thay đổi thư mục hiện tại sang một thư mục khác.
    
    - `rmdir` - loại bỏ toàn bộ thư mục. 

#### **Printing file and string contents**

- Đối với việc in nội dung tệp tin hoặc chuỗi, các lệnh phổ biến bao gồm: 

    - `cat` - in toàn bộ nội dung của một tệp tin.
    
    - `more` - được sử dụng để in nội dung tệp tin một trang một lần.
    
    - `head` - để in chỉ 'N' dòng đầu tiên của một tệp tin.
    
    - `tail` - để in chỉ 'N' dòng cuối cùng của một tệp tin.
    
    - `echo` - rất phổ biến - nó '`echoes`' một chuỗi đầu vào bằng cách in nó.

#### **Compression and archiving**

- Nếu muốn nén tệp tin hoặc ứng dụng lưu trữ, các lệnh shell liên quan bao gồm: 

    - `tar` - được sử dụng để lưu trữ một bộ các tệp tin.
    
    - `zip` - nén một bộ các tệp tin.
    
    - `unzip` - trích xuất các tệp tin từ một bộ lưu trữ được nén hoặc zip.

#### **Networking**

- Các ứng dụng mạng bao gồm các lệnh sau: 

    - `hostname` - in ra tên máy chủ.
    
    - `ping` - gửi gói tin đến một URL và in ra phản hồi.
    
    - `ifconfig` - hiển thị hoặc cấu hình giao diện mạng trên hệ thống.
    
    - `curl` - hiển thị nội dung của một tệp tin tại một URL.
    
    - `wget` - có thể được sử dụng để tải xuống một tệp tin từ một URL.

#### **Running Linux on a Windows machine**

- Chúng tôi nên đề cập rằng nếu bạn đang chạy máy tính chạy Windows và muốn chạy Linux trên đó, bạn có thể làm như vậy theo nhiều cách khác nhau. 

- Linux có thể được cài đặt trên một phân vùng ổ đĩa riêng. Chuyển đổi giữa hai hệ điều hành yêu cầu khởi động lại. 

- Hoặc bạn có thể cài đặt Linux trên máy ảo, hoặc bạn có thể cài đặt một bộ mô phỏng Linux như CygWin, hoặc sử dụng Windows Subsystem for Linux, một lớp tương thích để chạy các tệp thực thi nhị phân Linux trực tiếp trên Windows.

#### **Recap**

- Một shell là một giao diện tương tác để chạy lệnh, một ngôn ngữ kịch bản và một ngôn ngữ tương tác. 

- Các lệnh shell được sử dụng để duyệt và **làm việc với tệp tin và thư mục**. 

- Các lệnh shell có thể được **sử dụng để nén tệp tin**. 

- Các lệnh `curl` và `wget`, tương ứng, có thể được **sử dụng để hiển thị và tải xuống tệp tin từ URL**. 

- Lệnh `echo` **in chuỗi hoặc giá trị biến**. 

- Các lệnh `cat` và `tail` được sử dụng để **hiển thị nội dung tệp tin**.
