### **Informational Commands**

#### **User information**

- Bạn có thể sử dụng các lệnh thông tin trong terminal để **hiển thị thông tin người dùng về hệ thống của bạn**. 

- Các lệnh thông tin người dùng hữu ích khi bạn cần xác minh danh tính của người dùng hiện tại hoặc xác định tài khoản người dùng nào đang chạy một lệnh hoặc quy trình cụ thể. 

- Ví dụ, lệnh "`whoami`" **hiển thị tên người dùng hiện tại**. 

- Lệnh "`whoami`" **không yêu cầu đối số và không có tùy chọn**. 

- Dưới đây là một ví dụ về cách sử dụng khi ai đó đăng nhập với tên người dùng "`John Doe`". 

- Tương tự, lệnh "`id`" **trả về ID người dùng hoặc nhóm**, là **một số được gán cho mỗi người dùng hoặc nhóm trong hệ thống Linux**. 

- Trong ví dụ này, sử dụng "`id`" với tùy chọn "`minus u ( -u )`" **trả về ID số của người dùng**. 

- Nếu bạn muốn tên tương ứng với ID người dùng số, thêm tùy chọn "`minus n ( -n )`".

#### **System information**

- Lệnh "`uname`", viết tắt của "`Unix name`", **trả về thông tin về hệ điều hành như tên kernel và số phiên bản**. 

- Điều này có thể được sử dụng để **xác định loại hệ thống bạn đang làm việc** trên hoặc **chẩn đoán các vấn đề liên quan đến hệ thống**. 

- Nhập lệnh "`uname`" **trả về tên của hệ điều hành**, trong **trường hợp này là Darwin**. 

- Bao gồm các tùy chọn "`minus s ( -s )`" và "`minus r ( -r )`" **trả về cả tên hệ điều hành và phiên bản của nó**. 

- Bạn cũng có thể xem **thông tin phiên bản chi tiết hơn** bằng cách sử dụng tùy chọn "`minus v ( -v )`".

#### **Displaying your disk usage**

- Bạn có thể sử dụng lệnh "`df`", hoặc "`disk free`", để **hiển thị sử dụng ổ đĩa của hệ thống của bạn**. 

- Nó có ích trong các **tình huống bạn cần giám sát sử dụng ổ đĩa hoặc kiểm tra không gian khả dụng trên một hệ thống tệp cụ thể**. 

- Ví dụ, nhập "`df minus h tilde ( df -h ~ )`" hiển thị bảng sau, đặc biệt cho thư mục nhà của bạn, được biểu thị bằng ký hiệu `tilde ( ~ )`. 

- Trong bảng này, bạn có thể thấy tất cả các ổ đĩa được gắn trên thư mục nhà của bạn. 

- Trong Linux, bạn có thể "`mount`" một ổ đĩa vào một thư mục, **điều này có nghĩa là hệ thống tệp của ổ đĩa đó trở nên truy cập được thông qua thư mục đó**. 

- Bảng cũng **hiển thị tỷ lệ sử dụng bộ nhớ trên mỗi ổ đĩa**. 

- Tùy chọn "`minus h ( -h )`" giúp đưa ra đầu ra "dễ đọc cho người", biểu thị không gian ổ đĩa bằng đơn vị như Gigabyte và Terabyte thay vì Bytes.

- Để **xem sử dụng ổ đĩa trên tất cả các hệ thống tệp**, bạn chỉ cần gõ "`df minus h ( df -h )`" mà không cần chỉ định một thư mục. 

- Kết quả bao gồm kích thước, dung lượng đã sử dụng và không gian khả dụng cho mỗi hệ thống tệp.

#### **Displaying current running processes**

- Để xem các quy trình đang chạy trên hệ thống của bạn, bạn có thể sử dụng lệnh "`ps`", hoặc "`process status ( trạng thái quy trình )`". 

- Điều này hữu ích khi bạn cần giám sát hoặc quản lý quy trình. 

- Sử dụng "`ps`" với tùy chọn "`minus e ( -e )`" sẽ **liệt kê tất cả các quy trình đang chạy trên hệ thống, không phụ thuộc vào người dùng nào đã bắt đầu chúng**. 

- Lệnh "`ps`" **hiển thị thông tin như tên của mỗi quy trình đang chạy**, **ID quy trình và thời gian theo phút và giây mà mỗi quy trình đã chạy**.

#### **Monitoring system health and status**

- Lệnh "`top`", hoặc "`table of processes ( bảng quy trình )`", hoạt động như một **quản lý công việc và sẽ hiển thị một bảng các quy trình đang chạy và tài nguyên mà chúng đang sử dụng**. 

- Lệnh này **hữu ích khi bạn cần giám sát hiệu suất hệ thống** hoặc **xác định các quy trình tốn tài nguyên**. 

- Ở đây, chúng tôi hiển thị một ví dụ về việc sử dụng với tùy chọn "`minus n ( -n )`" và số "`3`" để **hiển thị ba nhiệm vụ đang chạy hàng đầu**: Chrome, top và Spotify. 

- **Mặc định, các nhiệm vụ được sắp xếp theo sử dụng CPU**. 

- Đây là đầu ra đơn giản hóa của lệnh "top" trong video này, nhưng "top" cung cấp nhiều thông tin khác nhau như sử dụng bộ nhớ và vị trí tệp thực thi.

#### **Printing strings and variable values**

- Mặc dù đơn giản, lệnh "`echo`" trong Linux là **một công cụ mạnh mẽ để hiển thị văn bản** hoặc **biến trên terminal hoặc trong một tập lệnh shell**. 

- Nhập chỉ "`echo`" **tương đương với việc nói với terminal không in gì cả, và nó sẽ trả về một dòng mới**. 

- Nếu bạn muốn **echo một từ duy nhất như "hello"**, bạn chỉ cần gõ "`echo hello`", và terminal sẽ trả lại từ đó. 

- Nói một cách chặt chẽ, bạn không cần thêm dấu ngoặc kép xung quanh một chuỗi có khoảng trắng để echo hoạt động như mong đợi, nhưng thường được coi là thực hành tốt để bao gồm dấu ngoặc kép. 

- **Sử dụng echo với một chuỗi trong dấu ngoặc kép trả về nội dung** được bao gồm trong ngoặc kép, `"Learning Linux is fun!"`. 

- Ngoài ra, bạn có thể **xem giá trị của một biến**, chẳng hạn như biến "`PATH`" của hệ thống, bằng cách **gõ dấu đô kèm theo tên biến**. 

- Điều này có thể **hữu ích khi gỡ lỗi hoặc viết tập lệnh**. Ở đây bạn **có thể thấy mỗi đường dẫn của biến PATH của hệ thống bạn, phân tách bởi dấu hai chấm**.

#### **Getting date information**

- Một lệnh khác hữu ích là "`date`", hiển thị ngày và giờ hệ thống hiện tại. 

- Nhập "`date`" trả lại **định dạng ngày mặc định: ngày trong tuần, ngày, tháng, năm, giờ và múi giờ**. 

- Bạn cũng có thể trích xuất một số phần cụ thể của ngày để in, như trong ví dụ này. 

- Để **định dạng đầu ra**, **bọc trong dấu ngoặc kép một kết hợp văn bản** và **ký tự kiểm soát được tiếp đầu ngữ bằng một dấu "cộng ( + )"**. 

- Các **ký tự kiểm soát định dạng được chỉ định bằng ký hiệu "%"**. Trong trường hợp này, `"%j"` và `"%Y"` **xuất số ngày trong năm và năm, tương ứng**. **Lệnh này in "97", cho ngày thứ 97 của năm, từ "ngày thứ", và năm "2023"**. 

- Lưu ý cách `"%Y" được thay thế bằng năm 2023`. Cuối cùng, đây là một ví dụ khác để thấy cách bạn có thể kết hợp các điều khiển định dạng với văn bản để in chuỗi duy nhất. 

- Bằng cách kết hợp các điều khiển như `"percent A"`, `"percent j," và "percent Y"` với văn bản, bạn có thể **in ra một dòng mô tả ngày của tuần, ngày và năm**.

- `%j` là số ngày trong một năm.

- `%A` là thứ.

- `%Y` là year.

- `%d` là day.

- `%B` là month

#### **Viewing the manual**

-  Nếu bạn muốn tìm hiểu thêm về cách sử dụng một lệnh, bạn có thể sử dụng lệnh `man`, hoặc "manual".

#### **RECAP**

**Video này chỉ đề cập đến những khái niệm cơ bản của các lệnh này.**

1. **Nhận thông tin người dùng với các lệnh "whoami" và "id":**
   - Lệnh `whoami`: Hiển thị tên người dùng hiện tại.
   - Lệnh `id -u`: Trả về ID số của người dùng.
   - Lệnh `id -nu`: Trả về tên người dùng tương ứng với ID số.

2. **Nhận thông tin hệ điều hành bằng cách sử dụng lệnh "uname":**
   - Lệnh `uname`: Trả về thông tin hệ điều hành, kernel, và phiên bản.
   - Ví dụ sử dụng `uname -sr`: Hiển thị tên hệ điều hành và phiên bản.

3. **Kiểm tra sử dụng ổ đĩa hệ thống bằng lệnh "df":**
   - Lệnh `df`: Hiển thị thông tin sử dụng ổ đĩa.
   - Ví dụ sử dụng `df -h ~`: Hiển thị thông tin về ổ đĩa home, với đơn vị là GB và TB.

4. **Giám sát quy trình và sử dụng tài nguyên bằng "ps" và "top":**
   - Lệnh `ps -e`: Liệt kê tất cả các quy trình đang chạy.
   - Lệnh `top -n 3`: Hiển thị bảng các quy trình đang chạy và sử dụng tài nguyên, giới hạn 3 dòng.
   
5. **In chuỗi hoặc giá trị biến bằng "echo":**
   - Lệnh `echo`: Hiển thị văn bản hoặc giá trị biến.
   - Ví dụ sử dụng `echo "hello"`: Hiển thị từ "hello".
   - Ví dụ sử dụng `echo $PATH`: Hiển thị giá trị của biến PATH.

6. **In và trích xuất thông tin về ngày bằng lệnh "date":**
   - Lệnh `date`: Hiển thị ngày và giờ hệ thống.
   - Ví dụ sử dụng `date "+%A, %d %B %Y"`: Hiển thị ngày theo định dạng thứ, ngày, tháng và năm.

7. **Đọc sách hướng dẫn cho bất kỳ lệnh nào bằng cách sử dụng "man":**
   - Lệnh `man`: Hiển thị sách hướng dẫn cho một lệnh cụ thể.
   - Ví dụ sử dụng `man ID`: Hiển thị sách hướng dẫn cho lệnh "ID", bao gồm cả tùy chọn và tham số.
   - Sử dụng `man man` để tìm hiểu thêm về lệnh "man" và cách sử dụng nó.
