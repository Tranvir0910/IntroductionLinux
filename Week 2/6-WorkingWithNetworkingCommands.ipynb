### **Working with Networking Commands**

#### **Display your system's hostname and IP address**

```python
$ hostname

$ hostname -i
```

`hostname` - hostname là tên được gán cho máy tính hoặc thiết bị trên mạng và được sử dụng để nhận dạng và liên lạc với thiết bị đó.

`-i` - để xem địa chỉ IP của máy chủ.

#### **Display network interface configuration**

```python
$ ifconfig

$ ifconfig eth0
```

`ifconfig` - được sử dụng để cấu hình hoặc hiển thị các tham số giao diện mạng cho mạng.

`ifconfig eth0` - hiển thị cấu hình của một thiết bị cụ thể, chẳng hạn như ethenet adapter eth0.

- Có thể thấy địa chỉ IP máy chủ của mình ở dòng 2 sau từ inet.

#### **Test connectivity to a host**

```python
$ ping www.google.com

$ ping -c 5 www.google.com
```

`ping` - Sử dụng lệnh ping để kiểm tra xem www.google.com có ​​thể truy cập được không. 

- Lệnh tiếp tục ping các gói dữ liệu đến máy chủ tại www.google.com và in phản hồi mà nó nhận được. (Nhấn Ctrl+c để dừng ping.)

- Muốn ping một số lần giới hạn, hãy sử dụng tùy chọn `-c`.

#### **Transfer data from a server**

```python
$ curl https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DB0250EN-SkillsNetwork/labs/Bash%20Scripting/usdoi.txt

$ curl -O https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DB0250EN-SkillsNetwork/labs/Bash%20Scripting/usdoi.txt
```

`curl` - Để truy cập tệp tại URL sau và hiển thị nội dung của tệp trên màn hình.

`curl -O` - truy cập tệp tại URL đã cho và cũng lưu nó vào thư mục làm việc hiện tại.

#### **Download file(s) from a URL**

```python
$ rm usdoi.txt

$ wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DB0250EN-SkillsNetwork/labs/Bash%20Scripting/usdoi.txt
```

- `wget` tương tự như lệnh `curl`, tuy nhiên công dụng chính của nó là **tải tập tin xuống**. 

- Một tính năng độc đáo của wget là nó có thể tải xuống các tệp đệ quy tại một URL.
