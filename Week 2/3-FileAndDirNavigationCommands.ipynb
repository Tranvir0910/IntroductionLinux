### **File and Directory Navigation Commands**

- Lệnh `ls` hoặc `'list'` sẽ **liệt kê các tệp và thư mục trong một thư mục**. 

- Nếu bạn nhập `'ls' trong thư mục home`, bạn sẽ **thấy tất cả các tệp và thư mục mà thư mục home của bạn chứa**. 

- Bạn cũng có thể truyền tên thư mục, chẳng hạn như 'Downloads', như một tham số cho ls, sẽ liệt kê nội dung của thư mục Downloads của bạn. 

- Lệnh ls cũng **hỗ trợ các tùy chọn để liệt kê thông tin bổ sung**. Giả sử bạn đang làm việc trong thư mục tài liệu của mình và bạn muốn biết thêm thông tin về các tệp trong thư mục. 

- Bạn có thể sử dụng lệnh `ls với tùy chọn -l`, sẽ **hiển thị các tệp và thư mục con dưới dạng một bảng chi tiết hơn**. 

- Bạn có thể thấy rằng terminal đã liệt kê tất cả các tệp và thư mục con, cùng với các chi tiết bổ sung như quyền, ngày sửa đổi cuối cùng và chủ sở hữu.

`ls /bin/b*` : bắt đầu bằng b.

`ls /bin/*r` : kết thúc bằng r

- Để có danh sách dài tất cả các tệp trong /etc:

`ls -la /etc`

- Kết hợp các tùy chọn -l và -a bằng cách sử dụng ký hiệu ngắn hơn, -la.

#### **Finding your working directory**

- Đôi khi, bạn có thể cần biết bạn đang làm việc trong thư mục nào. Trong những trường hợp như vậy, bạn có thể sử dụng lệnh `'print woring directory'` để biết thư mục làm việc hiện tại của bạn. 

- Nhập `"pwd"` vào dòng lệnh để sử dụng lệnh này. Ở đây, bạn có thể thấy rằng bạn đang làm việc trong thư mục home của bạn, `"/Users/me."`

#### **Navigating your directories**

- Nếu bạn muốn thay đổi thư mục làm việc hiện tại của mình, bạn có thể sử dụng lệnh `'change directory'`. 

- Lệnh `'change directory'`, hoặc `cd`, được sử dụng để **thay đổi thư mục bạn đang làm việc**. 

- Giả sử bạn đang ở trong thư mục home và muốn chuyển đến một thư mục con bên trong nó, chẳng hạn như thư mục `Documents`. 

- Đơn giản nhập `"cd Documents"` để thay đổi thư mục làm việc của bạn. 

- Bây giờ, nếu bạn nhập `"pwd"`, bạn có thể thấy rằng bạn đang làm việc trong thư mục con Documents của bạn. 

- Lệnh cd cho phép bạn chuyển đến các thư mục với cả đường dẫn tương đối và đường dẫn tuyệt đối. 

#### **Relative and absolute navigation**

- Giả sử bạn đang ở trong thư mục `"Notes"` trong cây thư mục `"Documents"` của bạn. 

- Để chuyển đến thư mục cha tương đối với thư mục hiện tại của bạn, nhập lệnh cd với biểu tượng đường dẫn tương đối ".." như là đối số của nó. 

- Bây giờ, thư mục làm việc của bạn là `"Math"` thư mục cha ngay sau thư mục `"Notes"` của bạn. 

- Nếu bạn muốn điều hướng trực tiếp đến thư mục home của bạn, hãy sử dụng ký hiệu `tilde ( ~ )` với lệnh cd. 

- Điều này sẽ đưa bạn đến thư mục home của bạn. Trong trường hợp này, ký hiệu `tilde ( ~ )` đại diện cho một đường dẫn tuyệt đối đến thư mục home. 

- Cuối cùng, bạn cũng có thể cung cấp một đường dẫn đầy đủ cho một thư mục. 

- Ở đây, bạn đang tham chiếu đến đường dẫn tuyệt đối cho thư mục `"Notes"` ban đầu. Như dự kiến, lệnh này sẽ đưa bạn trở lại thư mục `"Notes"` nơi bạn bắt đầu.

#### **Finding files**

- Cuối cùng, lệnh `'find'` là một **công cụ mạnh mẽ sẽ trả về đường dẫn của mọi tệp phù hợp với một điều kiện do người dùng chỉ định**. 

- Giả sử thư mục `"Documents"` của bạn có một cấu trúc tệp như hình, hai thư mục con, mỗi thư mục chứa một số tệp. 

- Hãy nói rằng bạn đang làm việc trong thư mục "Documents" của mình và muốn tìm các đường dẫn của tất cả các tệp có tên "a.txt" trong thư mục làm việc của bạn. 

- Để làm điều này, nhập `'find . -name 'a.txt'` `'. Đối số "."` có nghĩa là `"tìm kiếm trong đây,"` vì vậy lệnh **chỉ sẽ tìm kiếm trong thư mục làm việc hiện tại của bạn.**

- Để thực hiện một phiên bản **không phân biệt chữ hoa/thường** trong tìm kiếm của bạn, bạn có thể sử dụng tùy chọn `"-iname"` thay vì "-name". 

- Bạn sẽ thấy rằng bạn tìm thấy cùng một tệp, cộng thêm một tệp khác cùng tên nhưng có chữ cái hoa "A".

- lệnh sau tìm tất cả các tệp .txt trong thư mục /etc và tất cả các thư mục con của nó:

`find /etc -name \'*.txt\'`

- Lưu ý: Cùng với việc liệt kê tất cả các tệp .txt, thiết bị đầu cuối có thể trả về lỗi "Quyền bị từ chối". 

- Những lỗi này là bình thường vì bạn có quyền truy cập hạn chế trên máy thí nghiệm.

#### **RECAP**

- Lệnh ls liệt kê tất cả các tệp và thư mục trong một cây thư mục cụ thể. 

- Lệnh cd được sử dụng để điều hướng giữa các thư mục. 

- Đường dẫn tương đối liên quan đến thư mục làm việc hiện tại của bạn, trong khi đường dẫn tuyệt đối tự đứng độc lập. 

- Lệnh find được sử dụng để tìm kiếm tệp trong các thư mục của bạn.
