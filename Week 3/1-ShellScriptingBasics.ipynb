### **Introduction to Shell Scripting**

#### **What is a script?**

- Script là một danh sách các lệnh có thể được giải thích và chạy bởi một chương trình gọi là scripting language. 

- Các lệnh có thể được nhập tương tác tại dòng lệnh hoặc được liệt kê từng dòng trong một tệp văn bản. 

- scripting language thường không được biên dịch. Chúng được giải thích tại thời điểm chạy. 

- Script thường chậm hơn khi chạy so với các ngôn ngữ đã biên dịch, nhưng chúng cũng dễ dàng và nhanh chóng phát triển hơn nhiều.
